## Operacje na plikach

Zanim przejdziemy do operacji musimy zrozumieć czym jest plik w rozumieniu komputera. 
Plikt to "pojemnik" w komputerze do przechowywania danych.

Kiedy chcemy odczytać lub zapisać plik, musimy go najpierw otworzyć. Kiedy skończymy, należy go zamknąć, aby zwolnić zasoby powiązane z plikiem.

Dlatego w Pythonie operacja na plikach odbywa się w następującej kolejności:
1. Otwórz plik

### Zamykanie plików

Kiedy zakończymy wykonywanie operacji na pliku, musimy poprawnie go zamknąć.

Zamknięcie pliku zwolni zasoby, które były powiązane z plikiem. Odbywa się to za pomocą metody close() w Pythonie.


````python
file1 = open("test.txt", "r")

# odczyt
read_content = file1.read()
print(read_content)

# zamknięcie
file1.close()
````

Tutaj użyliśmy metody close() do zamknięcia pliku.

Po wykonaniu operacji na pliku zawsze powinniśmy zamknąć plik; to dobra praktyka programistyczna.

### Obsługa wyjątków w plikach
Jeśli pracujemy z otwartym plikiem i wystąpi wyjątek, kod zakończy działanie bez zamykania pliku. Bezpieczniejszym sposobem jest użycie bloku try… finally.

````python
try:
    file1 = open("test.txt", "r")
    read_content = file1.read()
    print(read_content)

finally:
    # zamknięcie
    file1.close()
````

Dzięki temu nawet przy wystąpieniu problemów zasoby zostaną zwolnione.

### Składnia with...

W Pythonie składnia with... pozwala nam na pracę z otwartym plikiem tylko w obrębie bloku kodu. Po wyjściu plik jest automatycznie zamykany.

````python
with open("test.txt", "r") as file1:
    read_content = file1.read()
    print(read_content)
````

**WAŻNE:** Taka składnia zapewnia nam zamknięcie pliku i zmniejsza liczbę linijek kodu. Dlatego warto korzystać z tego rozwiązania 😀

## Obsługa wyjątków w Pythonie

### Składnia try...except
Blok try...except jest wykorzystywany do obsługi wyjątków. Składnia wygląda następująco:

```
try:
    # kod, który może spowodować wyjątek
except:
    # kod, który zostanie wywołany w przypadku wystąpienia wyjątku
```

Kod, który może wygenerować wyjątek umieściłyśmy wewnątrz bloku try. Po każdym bloku try następuje blok except.

Kiedy wystąpi wyjątek, jest on przechwytywany przez blok except. Bloku except nie może występować bez try.

````python
try:
    numerator = 10
    denominator = 0

    result = numerator/denominator

    print(result)
except:
    print("Error: Denominator cannot be 0.")
````

W naszym przykładzie próbujemy podzielić liczbę przez 0. Taki kod generuje wyjątek.

Aby obsłużyć wyjątek, umieściliśmy kod `result = numerator/denominator` wewnątrz bloku `try`. Teraz, gdy wystąpi wyjątek, reszta kodu wewnątrz bloku `try` jest pomijana.

Blok `except` przechwytuje wyjątek, a instrukcje w nim zawarte zostaną wykonane.

Jeśli żadna z instrukcji w bloku try nie generuje wyjątku, blok except jest pomijany.

### Przechwytywanie określonych wyjątków w Pythonie

Dla każdego bloku try może być zero lub więcej bloków except. Wiele bloków except pozwala nam obsługiwać każdy typ wyjątku inaczej.

Typ argumentu każdego bloku oprócz wskazuje typ wyjątku, który może być przez niego obsłużony. 

````python
try:
    
    even_numbers = [2,4,6,8]
    print(even_numbers[5])

except ZeroDivisionError:
    print("Denominator cannot be 0.")
    
except IndexError:
    print("Index Out of Bound.")

# Output: Index Out of Bound
````

W tym przykładzie tworzymy listę even_numbers.

Ponieważ indeksujemy od 0, ostatni element ma index 0. 

```
print(even_numbers[5])
```

W bloku powyżej próbujemy wyciągnąć wartość pod indeksem 5, a zatem dostaniemy IndexError.

W przypadku wystąpienia IndexError, wyjątek ZeroDivisionError zostanie pominięty. Tylko kod pod blokiem `except IndexError` zostanie wywołany. 


### Blok try z else 
Może się zdarzyć, że będziemy chcieli aby jakiś kod został wywołany jeżeli kod w bloku try wywoła się bez wyjątków.

W takim przypadku możemy użyć słowa else po bloku try. Jest to opcjonalne!


````python
# chcemy wyświetlić liczbę odwrotną do liczby parzystej

try:
    num = int(input("Enter a number: "))
    assert num % 2 == 0 # assert sprawdza warunek, jeśli nie zostanie spełniony dostaniemy AssertionError
except:
    print("Not an even number!")
else:
    reciprocal = 1/num
    print(reciprocal)
````

````python
# wprowadz 1, potem 4, a potem 0
````

Jeśli wrzucimy 0 to dostaniemy ZeroDivisionError ponieważ kod w bloku else nie jest objęty obsługą wyjątków.

### Konstrukcja try...finally

Blok finally zostanie wywołany zawsze, niezależnie od wystąpienia lub nie wystąpienia wyjątku. Jest on opcjonalny i może być tylko jeden w konstrukcji try.

````python
try:
    numerator = 10
    denominator = 0

    result = numerator/denominator

    print(result)
except:
    print("Error: Denominator cannot be 0.")
    
finally:
    print("This is finally block.")
````

Pomimo wystąpienia wyjątku i tak blok finally się wykonał.

## Własne wyjątki


### Definiowanie własnych wyjątków

W Pythonie możemy definiować własne wyjątki, tworząc nową klasę dziedziczącą po wbudowanej klasie Exception.

Składnia do definiowania włąsnych wyjątków:

```
class CustomError(Exception):
    ...
    pass

try:
   ...

except CustomError:
    ...
```

CustomError to wyjątek zdefiniowany przez użytkownika, który dziedziczy po klasie Exception.

*Uwaga*

*Kiedy piszemy duży program w Pythonie, dobrą praktyką jest umieszczanie wszystkich zdefiniowanych przez nas wyjątków w oddzielnym pliku.
Wiele standardowych modułów definiuje swoje wyjątki oddzielnie jako exceptions.py lub errors.py (przeważnie, ale nie zawsze).*

````python
#definijemy własny wyjątek
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18" # komentarz do naszego wyjątku
    pass


number = 18 # liczba do odgadnięcia

try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise InvalidAgeException # raise ma zgłosić nasz włąsny wyjątek
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print("Exception occurred: Invalid Age")
````

W powyższym przykładzie zdefiniowaliśmy niestandardowy wyjątek InvalidAgeException, tworząc nową klasę pochodną, która dziedziczy po wbudowanej klasie Exception.

Jeżeli `input_number` jest mniejszy niż 18, kod wygeneruje wyjątek.

Gdy wystąpi wyjątek, reszta kodu w bloku try jest pomijana.

Blok wyjątku przechwytuje zdefiniowany przez użytkownika wyjątek InvalidAgeException i wykonywane są instrukcje wewnątrz bloku wyjątku.

### Dostosowywanie klas wyjątków do potrzeb
Możemy dodatkowo dostosować klasę pochodną, aby przyjmować inne argumenty zgodnie z naszymi potrzebami.

````python
class SalaryNotInRangeError(Exception):
    """Exception raised for errors in the input salary.

    Attributes:
        salary -- input salary which caused the error
        message -- explanation of the error
    """

    def __init__(self, salary, message="Salary is not in (5000, 15000) range"):
        self.salary = salary
        self.message = message
        super().__init__(self.message)


salary = int(input("Enter salary amount: "))
if not 5000 < salary < 15000:
    raise SalaryNotInRangeError(salary)
````

W powyższym przykładzie przeciążyliśmy konstruktor klasy Exception, aby przyjmował również nasze niestandardowe argumenty: salary i message.

Następnie konstruktor nadrzędnej klasy Exception jest wywoływany ręcznie z argumentem self.message za pomocą funkcji super().

Niestandardowy atrybut self.salary jest zdefiniowany do późniejszego użycia.

Odziedziczona metoda __str__ klasy Exception jest następnie używana do wyświetlania odpowiedniego komunikatu, gdy zostanie zgłoszony błąd SalaryNotInRangeError.

````python
'''
Napisz program, który będzie kalkulatorem:
- program ma pobierać od użytkownika input: liczba operator liczba
- każdy znak w inpucie po spacji, np. 1 + 1
- pobrany input należy przerobić na listę i sprawdzić jej poprawność:
  - jeżeli w liście jest więcej niż 3 elementy - podnieś wyjątek IllegalFormulaError
  - IllegalFormulaError ma być własnym wyjątkiem stworzonym przez Ciebie
  - zamień liczby na typ float
  - jeśli liczby zmieniane na float zwrócą ValueError - przechwyć go i podnieć IllegalFormulaError i wyświetl komunikat "1 i 3 wartość muszą być liczbam"
  - jeśli użytkownik zamiast +, -, * lub / da inny operator, również zwróć IllegalFormulaError
- jeśli nie wystąpił żaden wyjątek wyświetl wynik
- pobieraj input od użytkownika tak długo jak użytkownik nie wpisze w inpucike polecenia KONIEC
- działania na operatorach zawrzyj w funkcji

'''

class IllegalFormulaError(Exception):
    pass

def calculate(expression):
    if len(expression) != 3:
        raise IllegalFormulaError("Illegal formula: input musi mieć dokładnie 3 elementy")
    try:
        num1 = float(expression[0])
        num2 = float(expression[2])
    except ValueError:
        raise IllegalFormulaError("1 i 3 wartość muszą być liczbam")
    operator = expression[1]
    if operator not in ['+', '-', '*', '/']:
        raise IllegalFormulaError("Illegal operator")
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        if num2 == 0:
            raise IllegalFormulaError("Division by zero")
        return num1 / num2

while True:
    user_input = input("Podaj działanie (liczba operator liczba) lub KONIEC, aby zakończyć: ")
    if user_input.strip().upper() == 'KONIEC':
        break
    parts = user_input.strip().split()
    try:
        result = calculate(parts)
        print(f"Wynik: {result}")
    except IllegalFormulaError as e:
        print(f"Błąd: {e}")
````
