### Dziedziczenie
Dziedziczenie to sposób tworzenia nowej klasy z wykorzystaniem szczegółów istniejącej klasy bez konieczności jej modyfikowania.

Nowo utworzona klasa jest klasą pochodną (lub klasą potomną). Podobnie istniejąca klasa jest klasą bazową (lub klasą nadrzędną).

````python
# klasa bazowa
class Animal:

    def eat(self):
        print( "Lubie jesc!")

    def sleep(self):
        print("Lubie spac!")

# klasa pochodna
class Dog(Animal):

    def bark(self):
        print("Hau hau hau!!")

# Tworzymy obiekt klasy Dog
dog1 = Dog()

# Wywołujemy metody klasy bazowej
dog1.eat()
dog1.sleep()

# Wywołujemy metody klasy pochodnej
dog1.bark();
````

W tym przypadku dog1 (obiekt klasy pochodnej Dog) ma dostęp do członków klasy bazowej Animal. To dlatego, że Dog dziedziczy po Animal.

### Hermetyzacja (zwana również enkapsulacją)
Hermetyzacja jest jedną z kluczowych cech programowania obiektowego. Hermetyzacja odnosi się do łączenia atrybutów i metod w jednej klasie.

Uniemożliwia klasom zewnętrznym dostęp i zmianę atrybutów i metod klasy. Pomaga to również w ukryciu danych.

W Pythonie atrybuty chronione oznaczamy za pomocą podkreślenia jako przedrostka pojedynczego _  a prywatne podwójnego __. Na przykład:

````python
class Computer:

    def __init__(self):
        self.__maxprice = 900

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price

print("Tworzymy obiekt klasy z ceną:", end="\t")
c = Computer()
c.sell()

# zmieniamy cenę
print("Próbujemy nadpisać atrybut prywatny:", end="\t")
c.__maxprice = 1000
c.sell()

# zmieniamy cenę funkcją
print("Zmieniamy setterem:", end="\t")
c.setMaxPrice(1000)
c.sell()
````

W powyższym programie zdefiniowaliśmy klasę Computer.

Użyliśmy metody __init__() do przechowywania maksymalnej ceny sprzedaży komputera. Tutaj, zwróćmy uwagę na kod:

```
c.__maxprice = 1000
```

Próbowaliśmy zmodyfikować wartość __maxprice poza klasą. Ponieważ jednak __maxprice jest zmienną prywatną, ta modyfikacja nie jest widoczna na danych wyjściowych.

Jak widzimy, aby zmienić wartość, musimy użyć funkcji setter, tj. setMaxPrice(), która przyjmuje cenę jako parametr.

Inny przykład:

````python
class Circle:
    def __init__(self, radius):
        self.__radius = radius  # private variable

    # Getter method
    def get_radius(self):
        return self.__radius

    # Setter method
    def set_radius(self, radius):
        if radius >= 0:
            self.__radius = radius
        else:
            print("Radius cannot be negative!")

    def area(self):
        return 3.14 * self.__radius * self.__radius

# Testing the encapsulation

c = Circle(5)
print(c.area())      # prints the area

# Direct access to __radius will result in an error
# print(c.__radius)  # AttributeError

# Using getters and setters
print(c.get_radius())   # prints 5
c.set_radius(-10)       # prints "Radius cannot be negative!"
````

### Polimorfizm
Polimorfizm to kolejna ważna koncepcja programowania obiektowego. Oznacza po prostu więcej niż jedną formę.

Oznacza to, że ten sam byt (metoda, operator lub obiekt) może wykonywać różne operacje w różnych scenariuszach.

Rozważmy przykład,

````python
class Polygon:
    # metoda renderowania kształtu
    def render(self):
        print("Rendering Polygon...")

class Square(Polygon):
    # renderuje kwadrat
    def render(self):
        print("Rendering Square...")

class Circle(Polygon):
    # renderuje okrąg
    def render(self):
        print("Rendering Circle...")

# tworzymy obiekt Square
s1 = Square()
s1.render()

# tworzymy obiekt Circle
c1 = Circle()
c1.render()
````

W powyższym przykładzie stworzyliśmy klasę nadrzędną: Polygon (wielokąt) i dwie dziedziczące: Kwadrat i Okrąg. Zwróćmy uwagę na użycie metody render().

Głównym celem metody render() jest renderowanie kształtu. Jednak proces renderowania kwadratu różni się od procesu renderowania okręgu.

Dlatego metoda render() zachowuje się inaczej w różnych klasach. Lub możemy powiedzieć, że render() jest polimorficzny.

## Dziedziczenie

Podobnie jak inne języki OOP, Python obsługuje również koncepcję dziedziczenia klas.

Dziedziczenie pozwala nam stworzyć nową klasę z istniejącej klasy.

Nowo utworzona klasa jest nazywana podklasą (klasa podrzędna lub klasa pochodną), a istniejąca klasa, z której pochodzi klasa podrzędna, jest nazywana klasą nadrzędną (klasą bazową).



### Składnia dziedziczenia

Dziedziczenie odbywa się w następujący sposób:

```
# definijemy klasę nadrzedną
class super_class:
    # definicja metod i atrybutów

# dziedziczenie
class sub_class(super_class):
    # metody i atrybuty super_class # muszą być pierwsze w konstruktorze
    # metody i atrybuty sub_class

```
Tutaj klasa sub_class dziedziczy po super_class.

````python
# przykład

class Animal:

    # atrybut i metoda klasy nadrzędnej
    name = ""

    def eat(self):
        print("I can eat")

# dziedziczymy po Animal
class Dog(Animal):

    # nowa metoda w subclass
    def display(self):
        # dostęp do atrybutu name nadklasy za pomocą self
        print("My name is ", self.name)

# worzymy obiekt klasy dziedziczącej
labrador = Dog()

# dostęp do atrybutu i metody klasy nadrzędnej
labrador.name = "Rohu"
labrador.eat()

# wywołujemy metodę klasy dziedziczącej
labrador.display()
````

W powyższym przykładzie wyprowadziliśmy podklasę Dog z nadklasy Animal. Zwróćmy uwagę na instrukcje:
```
labrador.name = "Rohu"

labrador.eat()
```

Tutaj używamy labradora (obiekt Dog), aby uzyskać dostęp do name i eat() klasy Animal. Jest to możliwe, ponieważ podklasa dziedziczy wszystkie atrybuty i metody klasy nadrzędnej.

Ponadto uzyskaliśmy dostęp do atrybutu name wewnątrz metody klasy Dog za pomocą self.

*Przykład dziedziczenia*

Rzućmy okiem na inny przykład dziedziczenia:

Wielokąt to zamknięta figura, która ma 3 lub więcej boków. Powiedzmy, że mamy klasę o nazwie Polygon zdefiniowaną w następujący sposób:

````python
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
````

Ta klasa ma atrybuty do przechowywania liczby boków n i wielkości każdego boku w postaci listy zwanej sides (bokami).

* Metoda inputSides() przyjmuje wielkość każdego boku
* Metoda dispSides() wyświetla długości boków


Trójkąt to wielokąt o 3 bokach. Możemy więc stworzyć klasę o nazwie Triangle, która dziedziczy po Polygon. Dzięki temu wszystkie atrybuty klasy Polygon są dostępne dla klasy Triangle.

Nie musimy ich ponownie definiować (możliwość ponownego użycia kodu). Trójkąt można zdefiniować w następujący sposób.

````python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # obliczamy połowę obwodu
        s = (a + b + c) / 2
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
````

Jednak klasa Triangle ma nową metodę findArea() służącą do znajdowania i drukowania pola trójkąta.

Zobaczmy teraz pełny działający kod z powyższego przykładu, w tym utworzenie obiektu,

````python
class Polygon:
    # IInicjalizujemy liczbę boków
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    # metoda wyświetlania długości każdego boku wielokąta
    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])

class Triangle(Polygon):
    # Zainicjowanie liczby boków trójkąta do 3
    # przez wywołanie metody __init__ klasy Polygon
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides

        # obbliczenie połowy obwodu
        s = (a + b + c) / 2

        # Wykorzystanie wzoru Herona do obliczenia pola trójkąta
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)

# Tworzenie instancji klasy Triangle
t = Triangle()

# Użytkownik zostaje poprpszony o wprowadzenie boków trójkąta
t.inputSides()

# Wyświetlanie boków trójkąta
t.dispSides()

# Obliczanie i wyświetlanie pola trójkąta
t.findArea()
````

Tutaj widzimy, że chociaż nie zdefiniowaliśmy osobno metod takich jak inputSides() czy dispSides() dla klasy Triangle, mogliśmy z nich korzystać.

Jeśli atrybut nie zostanie znaleziony w samej klasie, wyszukiwanie jest kontynuowane w klasie bazowej. Powtarza się to rekurencyjnie, jeśli sama klasa bazowa pochodzi od innych klas.

### Przesłanianie metod w dziedziczeniu

W poprzednim przykładzie zobaczyliśmy, że obiekt podklasy może uzyskać dostęp do metody nadklasy.

Co jednak, jeśli ta sama metoda jest obecna zarówno w nadklasie, jak i podklasie?

W tym przypadku metoda w podklasie zastępuje metodę w nadklasie. Ta koncepcja jest znana jako przesłanianie metod w Pythonie.

````python
class Animal:

    # aatrybuty i metody klasy nadrzędnej
    name = ""

    def eat(self):
        print("I can eat")

# dziedziczymy po Animal
class Dog(Animal):

    # przesłaniamy metodę eat()
    def eat(self):
        print("I like to eat bones")

# tworzymy obiekt klasy podrzędnej
labrador = Dog()

# wywołujemy metodę eat() na rzecz obiektu labrador
labrador.eat()
````

W powyższym przykładzie ta sama metoda eat() występuje zarówno w klasie Dog, jak i w klasie Animal.

Teraz, kiedy wywołujemy metodę eat() przy użyciu obiektu podklasy Dog, wywoływana jest metoda klasy Dog.

Dzieje się tak, ponieważ metoda eat() klasy podrzędnej Dog zastępuje tę samą metodę klasy nadrzędnej Animal.

### Metoda super() w dziedziczeniu
Wcześniej widzieliśmy, że ta sama metoda w podklasie zastępuje metodę w nadklasie.

Jeśli jednak potrzebujemy uzyskać dostęp do metody nadklasy z podklasy, używamy metody super(). Na przykład,

````python
class Animal:

    name = ""

    def eat(self):
        print("I can eat")

# dziedziczy po Animal
class Dog(Animal):

    # przesłania metodę eat()
    def eat(self):

        # wywołuje metodę eat() superclass dzięki super()
        super().eat()

        print("I like to eat bones")

# instancja klasy Dog
labrador = Dog()

labrador.eat()
````

W powyższym przykładzie metoda eat() klasy Dog zastępuje tę samą metodę klasy nadrzędnej Animal.

W klasie Dog wywołaliśmy metodę eat() klasy Animal.

```
# wywołanie metody superclass
super().eat()
```

Zatem jeśli wywołamy metodę eat() na obiekcie labrador
```
# wywołąnie metody eat()
labrador.eat()
```
Wykonywana jest zarówno przesłonięta, jak i nadklasowa wersja metody eat() .

## Wielodziedziczenie

Klasa może pochodzić z więcej niż jednej nadklasy w Pythonie. Nazywa się to wielokrotnym dziedziczeniem.

Na przykład klasa Bat wywodzi się z nadklas Mammal i WingedAnimal. Ma to sens, ponieważ nietoperz jest zarówno ssakiem, jak i skrzydlatym zwierzęciem.

image.png

### Składnia wielodziedziczenia

```
class SuperClass1:
    # cechy SuperClass1

class SuperClass2:
    # cechy SuperClass2

class MultiDerived(SuperClass1, SuperClass2):
    # cechy SuperClass1 + SuperClass2 + MultiDerived
```
Tutaj klasa MultiDerived pochodzi od klas SuperClass1 i SuperClass2.

````python
# przykład

class Mammal:
    def mammal_info(self):
        print("I am mammal.")

class WingedAnimal:
    def winged_animal_info(self):
        print("I have wings.")

class Bat(Mammal, WingedAnimal):
    pass

# tworzymy obiekt klasy Bat
b1 = Bat()

b1.mammal_info()
b1.winged_animal_info()
````

W powyższym przykładzie klasa Bat wywodzi się z dwóch superklas: Mammal i WingedAnimal. Zwróćmy uwagę na instrukcję:
```
b1 = Bat()
b1.mammal_info()
b1.winged_animal_info()
```
Tutaj używamy b1 (obiekt Bat), aby uzyskać dostęp do metod ssak_info() i winged_animal_info() odpowiednio klasy Mammal i WingedAnimal.



### Dziedziczenie wielopoziomowe

W Pythonie możemy nie tylko wyprowadzić klasę z klasy nadrzędnej, ale także wyprowadzić klasę z klasy pochodnej. Ta forma dziedziczenia jest znana jako dziedziczenie wielopoziomowe.

Oto składnia wielopoziomowego dziedziczenia:

````python
class SuperClass:
    # kod klasy SuperClass

class DerivedClass1(BaseClass):
    # kod klasy DerivedClass1 - 1 poziom dziedziczenia

class DerivedClass2(DerivedClass1):
    # kod klasy DerivedClass2 - 2 poziom dziedziczenia
````

Tutaj klasa DerivedClass1 pochodzi od klasy SuperClass, a klasa DerivedClass2 pochodzi od klasy DerivedClass1.

image.png

````python
# Przykłąd dziedziczenia wielopoziomowego
class SuperClass:

    def super_method(self):
        print("Super Class method called")

# definijemy klasę dziedziczącą po SuperClass
class DerivedClass1(SuperClass):
    def derived1_method(self):
        print("Derived class 1 method called")

# definijemy klasę dziedziczącą po DerivedClass1
class DerivedClass2(DerivedClass1):

    def derived2_method(self):
        print("Derived class 2 method called")

# tworzymy obiekt klasy DerivedClass2
d2 = DerivedClass2()

d2.super_method()  # Output: "Super Class method called"

d2.derived1_method()  # Output: "Derived class 1 method called"

d2.derived2_method()  # Output: "Derived class 2 method called"
````

W powyższym przykładzie DerivedClass2 pochodzi od DerivedClass1, która pochodzi od SuperClass.

Oznacza to, że DerivedClass2 dziedziczy wszystkie atrybuty i metody zarówno DerivedClass1, jak i SuperClass.

Dlatego używamy d2 (obiekt DerivedClass2) do wywoływania metod z SuperClass, DerivedClass1 i DerivedClass2.

### Zasady Wyszukowania Metod w Pythonie (MRO - Method Resolution Order)
Jeśli dwie nadklasy mają tę samą nazwę metody, a klasa pochodna wywołuje tę metodę, Python używa MRO do wyszukania właściwej metody do wywołania. Na przykład:

````python
class SuperClass1:
    def info(self):
        print("Super Class 1 method called")

class SuperClass2:
    def info(self):
        print("Super Class 2 method called")

class Derived(SuperClass1, SuperClass2):
    pass

d1 = Derived()
d1.info()

# Output: "Super Class 1 method called"
````

W naszym przykłądzie obie klasy SuperClass1 i SuperClass2  definiują metodę info().

Więc kiedy info() jest wywoływana przy użyciu obiektu d1 klasy Derived, Python używa MRO do określenia, którą metodę wywołać.

W tym przypadku MRO określa, że metody powinny być najpierw dziedziczone z nadklasy znajdującej się najbardziej na lewo, więc wywoływana jest funkcja info() klasy SuperClass1, a nie klasy SuperClass2.

## Przeciążanie operatorów

## Przykład przeicążania operatora

Aby przeciążyć operator +, będziemy musieli zaimplementować w klasie funkcję __add__().

Możemy zrobić cokolwiek zechcemy wewnątrz tej funkcji. Ale bardziej sensowna jest implementacja, która zwróci obiek Point na bazie sumy współdzędnych.

````python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x, self.y)

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)


p1 = Point(1, 2)
p2 = Point(2, 3)

print(p1+p2)

# Output: (3,5)
````

W powyższym przykładzie możemy zobaczyć, że kiedy używamy p1 + p2, Python wywołuje p1.__add__(p2), co z kolei odpowiada Point.__add__(p1,p2). Następnie operacja dodawania jest przeprowadzana w określony przez nas sposób.

### Przeciążanie operatorów do porównywania

Python nie ogranicza przeciążania operatorów tylko do operatorów arytmetycznych. Możemy również przeciążyć operatory porównania.

Poniżej przykład, w jaki sposób możemy przeciążyć operator <, żeby porównać ze sobą dwa obiekty klasy Person pod względem wieku:



````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # przeciążamy operator <
    def __lt__(self, other):
        return self.age < other.age

p1 = Person("Alice", 20)
p2 = Person("Bob", 30)

print(p1 < p2)  #  True
print(p2 < p1)  #  False
````

W tym przypadku __lt__() przeciąża operator < w celu porównania atrybutu wieku dwóch obiektów.

Metoda __lt__() zwraca:

* True - jeśli wiek pierwszego obiektu jest mniejszy niż wiek drugiego obiektu
* False — jeśli wiek pierwszego obiektu jest większy niż wiek drugiego obiektu
