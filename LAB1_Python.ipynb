### Instrukcje warunkowe

#### **if** ... **else**

Konstrukcja if ... else pozwala nam tworzyć instrukcje warunkowe - np. do rozwiązywania problemów gdzie należy podjąć jakąś decyzję.

W programowaniu korzystamy z polecenia *if* jeśli dany blok kodu ma zostać wywołany tylko w przypadku spełnienia warunku.

Np. wystawiam oceny studentom i automatyzuję sobie proces:


```
jeśli pkt jest ponad 90:
  przypisz_ocenę 5
jeśli pkt jest ponad 80:
  przypisz_ocenę 4
jeśli pkt jest ponad 70:
  przypisz_ocenę 3
w przeciwnym razie:
  przypisz_ocenę 2

```

W Pythonie rozróżniamy 3 wersje polecenia if ... else:
1. `if`
2. `if ... else`
3. `if ... elif ... else`



````python
'''
if warunek:
  blok kodu, który się wykona, jeśli warunkek będzie spełniony (wartość True)

blok kodu niezależny od warunku
'''
````

#### **match** ... **case**

Nowa struktura dostępna dopiero od Pythona 3.10. Może nam zastąpić if ... elif ... else

````python
# wersja podstawowa

#command = 'Hello, World!'
command = 'Goodbye, World!'
match command:
  case 'Hello, World!': # if albo elif
    print('Hello to you too!')
  case 'Goodbye, World!':
    print('See you later')
  case other: # działa jak else
    print('No match found')
````

````python
# korzystając z pakietu os można faktycznie wdrożyć takie usuwanie wybranych plików
def file_handler_v1(command):
    match command.split():
        case ['show']:
            print('List all files and directories: ')
            # wpisanie listy plików
        case ['remove', *files]:
            print('Removing files: {}'.format(files))
            # usunięcie plików
        case _:
            print('Command not recognized')
````

### Pętle

Pętle pozwalają wielokrotnie wykonać blok kodu. Mamy dwa typy: for i while

#### pętla **for**

Pozwala na wykonanie bloku kodu N razy. Zazwyczaj stosujemy ją do przedlądania kolekcji jak listy, krotki, napisy etc.

```
for wartosc in kolekcja:
  # jakis kod
```

image.png

````python
# pętla for z napisem
my_string = "Hello world"
for char in my_string:
  print(char)
````

````python
# pętla for z funkcją range
for i in range(10):
  print('hello')
  #print(i)
````

````python
# pętla for z kolekcjami
list_of_programming_langs = ['Python', 'Java', 'C++', 'JavaScript', 'C', 'Pearl']
for lang in list_of_programming_langs:
  print(lang)
````

````python
# pętla for z else

for item in ['computer', 'book', 'pen', 'phone']:
  print(item)
else:
  print("items' list is empty")
````

#### pętla **while**

Wykonuje się tak długo jak warunek jest spełniany,
ale najpierw sprawdza warunek, a dopiero potem wykonuje ciało pętli.

```
while warunek is True:
  # jakis kod
```

image.png

````python
# wyświetlimy liczby od 1 do 5

# inicjalizujemy zmienne
i = 1
n = 5

# pętla while loop dla i = 1, które zwiększamy do 5
while i <= n:
    print(i)
    i = i + 1
````

````python
# nieskończona pętla while

while True:
  # ciało pętli
````

````python
# pętla while z else
counter = 3

while counter > 0:
  print(f"wykonuje się część while, gdzie counter jest równy: {counter}")
  counter -= 1
else:
  print(f"counter jest równy: {counter}")
````

**WAŻNE: pętle for wybieramy, kiedy wiemy jaka ma być liczba iteracji, w przeciwnym razie bierzemy while.**

### *break* ... *continue*

Polecenia **break** używamy, gdy chcemy przerwać wykonywanie kodu, np. po spełnieniu jakiegoś warunku.
Poniższa

```
for val in sequence:
  # code
  if condition:
    break # przerywamy wykonywanie pętli for

while condition:
  # code
  if condition:
    break # przerywamy wykonywanie pętli while
```

````python
# z pętla for
for i in range(5):
    if i == 3:
        break
    print(i)
````

````python
# pętlą while
# szukamy pierwszych 5 wielokrotności 6

i = 1

while (i<=10):
    print('6 * ',i, '=',6 * i)

    if i >= 5:
        break

    i = i + 1
````

Polecenie **continue** pomija aktualną iterację pętli i przechodzi do kolejnej.

image.png

````python
# jaki będzie wynik?
for i in range(5):
    if i == 3:
        continue
    print(i)
````

````python
# szukamy liczb nieparzystych w przedziale 0-10

num = 0

while num < 10:
    num += 1

    if (num % 2) == 0:
        continue

    print(num)
````
