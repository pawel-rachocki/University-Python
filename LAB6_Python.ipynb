## Iteratory Pythona

Iteratory to metody, które iterują kolekcje, takie jak listy, krotki itp. Za pomocą metody iteratora możemy przechodzić przez obiekt i zwracać jego elementy.

Z technicznego punktu widzenia obiekt iteratora Pythona musi implementować dwie specjalne metody, __iter__() i __next__(), zwane łącznie protokołem iteratora.

Iteracja przez iterator
W Pythonie możemy użyć funkcji next(), aby zwrócić następny element w sekwencji.

````python
# define a list
my_list = [4, 7, 0]

# tworzymy iterator z listy
iterator = iter(my_list)

# pobieramy pierwszy element iteratora
print(next(iterator))

# pobieramy drugi element iteratora
print(next(iterator))

# pobieramy trzeci element iteratora
print(next(iterator))
````

Tutaj najpierw stworzyliśmy iterator z listy za pomocą metody iter(). Następnie użyłem funkcji next() do pobrania elementów iteratora w kolejności sekwencyjnej.

Kiedy dojdziemy do końca i nie będzie więcej danych do zwrócenia, otrzymamy wyjątek StopIteration.

### Korzystanie z pętli for
Bardziej eleganckim sposobem automatycznej iteracji jest użycie pętli for. Na przykład,

````python
# definiujemy listę
my_list = [4, 7, 0]

for element in my_list:
    print(element)
````

### Działanie pętli for dla iteratorów
Pętla for w Pythonie służy do iteracji po sekwencji elementów, takich jak lista, krotka lub ciąg znaków.

Kiedy używamy pętli for z iteratorem, pętla będzie automatycznie iterować po elementach iteratora, aż się wyczerpie.

Oto przykład działania pętli for z iteratorem,

````python
# tworzymy listę integerów
my_list = [1, 2, 3, 4, 5]

# przekształcamy listę w iterator
iterator = iter(my_list)

# iterujemy po elementach iteratora
for element in iterator:

    # wypisujemy element do konsoli
    print(element)
````

W tym przykładzie pętla for wykonuje iterację elementów obiektu iteratora.

W każdej iteracji pętla przypisuje wartość następnego elementu do elementu zmiennej, a następnie wykonuje wcięty blok kodu.

Ten proces trwa do wyczerpania iteratora, w którym to momencie kończy się pętla for.

### Tworzenie niestandardowych iteratorów
Budowanie iteratora od podstaw jest łatwe w Pythonie. Musimy tylko zaimplementować metody __iter__() i __next__(),

__iter__() zwraca sam obiekt iteratora. W razie potrzeby można przeprowadzić inicjalizację.
__next__() musi zwrócić następny element w sekwencji. Po dotarciu do końca i w kolejnych wywołaniach musi podnieść StopIteration.
Zobaczmy przykład, który da nam następną potęgę 2 w każdej iteracji. Wykładnik potęgi zaczyna się od zera do numeru zestawu użytkownika,

````python
class PowTwo:
    """Klasa do implementacji iteratora
     potęg dwójki"""

    def __init__(self, max=0):
        self.max = max

    def __iter__(self):
        self.n = 0
        return self

    def __next__(self):
        if self.n <= self.max:
            result = 2 ** self.n
            self.n += 1
            return result
        else:
            raise StopIteration


# tworzymy obiekt
numbers = PowTwo(3)

# tworzymy iterowalny obiekt
i = iter(numbers)

# używamy next, aby przejść do następnego elementu iteratora
print(next(i)) # printujemy 1
print(next(i)) # printujemy 2
print(next(i)) # printujemy 4
print(next(i)) # printujemy 8
print(next(i)) # podniesie wyjątek StopIteration
````

Możemy również użyć pętli for do iteracji po naszej klasie iteratora.

````python
for i in PowTwo(3):
    print(i)
````

### Nieskończone iteratory Pythona
Nieskończony iterator to iterator, który nigdy się nie kończy, co oznacza, że będzie nadal generował elementy w nieskończoność.

Oto przykład tworzenia nieskończonego iteratora w Pythonie za pomocą funkcji count() z modułu itertools,

````python
from itertools import count

# tworzymy nieskończony iterator, który zaczyna się od 1
# i zwiększa się o 1 za każdym razem
infinite_iterator = count(1)

# drukujemy pierwsze 5 elementów nieskończonego iteratora
for i in range(5):
    print(next(infinite_iterator))
````

Tutaj stworzyliśmy nieskończony iterator, który zaczyna się od 1 i za każdym razem zwiększa się o 1.

Następnie wydrukowaliśmy pierwszych 5 elementów nieskończonego iteratora za pomocą pętli for i metody next().

## Generatory Pythona

W Pythonie generator jest funkcją, która zwraca iterator, który tworzy sekwencję wartości podczas iteracji.

Generatory są przydatne, gdy chcemy wygenerować dużą sekwencję wartości, ale nie chcemy przechowywać ich wszystkich w pamięci naraz.

### Utwórz generator Pythona
W Pythonie, podobnie jak w przypadku normalnej funkcji, możemy zdefiniować funkcję generatora za pomocą słowa kluczowego def, ale zamiast instrukcji return używamy instrukcji yield.

```
def generator_name(args):
  some_statements
  yield result
```

````python
def generator_name(arg):
    # statements
    yield something
````

Tutaj słowo kluczowe yield jest używane do wygenerowania wartości z generatora.

Wywołanie funkcji generatora nie powoduje natychmiastowego wykonania treści funkcji. Zamiast tego zwraca obiekt generatora, który można iterować w celu wygenerowania wartości.

### Przykład: Generator Pythona
Oto przykład funkcji generatora, która tworzy sekwencję liczb,

````python
def my_generator(n):

    # inicjujemy licznik
    value = 0

    # wykonujemy pętlę, aż licznik będzie mniejszy niż n
    while value < n:

        # podajemy aktualną wartość licznika
        yield value

        # zwiększamy licznik
        value += 1

# iterujemy po obiekcie generatora utworzonym przez my_generator
for value in my_generator(3):

    # drukujemy każdą wartość wygenerowaną przez generator
    print(value)
````

W powyższym przykładzie funkcja generatora my_generator() przyjmuje jako argument liczbę całkowitą n i generuje sekwencję liczb od 0 do n-1.

Słowo kluczowe yield służy do generowania wartości z generatora i wstrzymywania wykonywania funkcji generatora do momentu zażądania następnej wartości.

Pętla for iteruje po obiekcie generatora utworzonym przez my_generator(), a instrukcja print wypisuje każdą wartość utworzoną przez generator.

### Użycie generatorów Pythona
Istnieje kilka powodów, dla których generatory są potężną implementacją.

1. **Łatwe do wdrożenia**

Generatory można implementować w przejrzysty i zwięzły sposób w porównaniu z ich odpowiednikami w klasie iteratorów. Poniżej znajduje się przykład implementacji sekwencji mocy 2 przy użyciu klasy iteratora.

````python
class PowTwo:
    def __init__(self, max=0):
        self.n = 0
        self.max = max

    def __iter__(self):
        return self

    def __next__(self):
        if self.n > self.max:
            raise StopIteration

        result = 2 ** self.n
        self.n += 1
        return result
````

Powyższy program był długi i mylący. Teraz zróbmy to samo, używając funkcji generatora.

````python
def PowTwoGen(max=0):
    n = 0
    while n < max:
        yield 2 ** n
        n += 1
````

Ponieważ generatory automatycznie śledzą szczegóły, implementacja była zwięzła i znacznie czystsza.

2. **Wydajna pamięć**

Normalna funkcja zwracająca sekwencję utworzy całą sekwencję w pamięci przed zwróceniem wyniku. Jest to przesada, jeśli liczba elementów w sekwencji jest bardzo duża.

Implementacja generatora takich sekwencji jest przyjazna dla pamięci i jest preferowana, ponieważ generuje tylko jeden element na raz.

3. **Reprezentuje nieskończony strumień**

Generatory są doskonałymi nośnikami do reprezentowania nieskończonego strumienia danych. Nieskończone strumienie nie mogą być przechowywane w pamięci, a ponieważ generatory wytwarzają tylko jeden element naraz, mogą reprezentować nieskończony strumień danych.

Poniższa funkcja generatora może generować wszystkie liczby parzyste (przynajmniej w teorii).

````python
def all_even():
    n = 0
    while True:
        yield n
        n += 2
````

4. **Generatory w pipline**

Do potokowania serii operacji można użyć wielu generatorów. Najlepiej zilustruje to przykład.

Załóżmy, że mamy generator, który generuje liczby w szeregu Fibonacciego. I mamy inny generator do kwadratu liczb.

Jeśli chcemy znaleźć sumę kwadratów liczb w szeregu Fibonacciego, możemy to zrobić w następujący sposób, łącząc potokowo wyjście funkcji generatora.

````python
def fibonacci_numbers(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2

print(sum(square(fibonacci_numbers(10))))
````

To potokowanie (pipelining) jest wydajne i łatwe do odczytania (i tak, dużo fajniejsze!).

## Domknięcia w Pythonie

Domknięcie Pythona to funkcja zagnieżdżona, która pozwala nam na dostęp do zmiennych funkcji zewnętrznej nawet po jej zamknięciu.

Zanim nauczymy się domknięcia, omówmy najpierw koncepcję funkcji zagnieżdżonych w Pythonie.

### Zagnieżdżona funkcja w Pythonie
W Pythonie możemy stworzyć funkcję wewnątrz innej funkcji. Jest to znane jako funkcja zagnieżdżona. Na przykład,

````python
def greet(name):
    # funkcja wewnętrzna
    def display_name():
        print("Hi", name)

    # wywołanie funkcji wewnętrznej
    display_name()

# wywołanie funkcji zewnetrznej
greet("John")
````

W powyższym przykładzie zdefiniowaliśmy funkcję display_name() wewnątrz funkcji greet().

Tutaj display_name() jest funkcją zagnieżdżoną. Funkcja zagnieżdżona działa podobnie do normalnej funkcji. Wykonuje się, gdy nazwa_wyświetlania() jest wywoływana wewnątrz funkcji witaj().

### Zamknięcia Pythona
Jak już omówiliśmy, zamknięcie jest funkcją zagnieżdżoną, która pomaga nam uzyskać dostęp do zmiennych funkcji zewnętrznej nawet po zamknięciu funkcji zewnętrznej. Na przykład,

````python
def greet():
    # zmienna zdefiniowana poza funkcją wewnętrzną
    name = "John"

    # zwracamy zagnieżdżoną funkcję anonimową
    return lambda: "Hi " + name

# wywołujemy funkcję zewnętrzną
message = greet()

# wywołujemy funkcję wewnętrzną
print(message())
````

W powyższym przykładzie utworzyliśmy funkcję o nazwie greet(), która zwraca zagnieżdżoną funkcję anonimową.

Tutaj, kiedy wywołujemy funkcję zewnętrzną,

```
message = greet()
```

Zwrócona funkcja jest teraz przypisana do zmiennej message.

W tym momencie wykonanie funkcji zewnętrznej jest zakończone, więc zmienna name powinna zostać zniszczona. Jednak gdy wywołujemy funkcję anonimową korzystając z

```
print(message())
```

jesteśmy w stanie uzyskać dostęp do zmiennej name funkcji zewnętrznej.

Jest to możliwe, ponieważ funkcja zagnieżdżona działa teraz jako zamknięcie, które zamyka zmienną zakresu zewnętrznego w swoim zakresie nawet po wykonaniu funkcji zewnętrznej.

Zobaczmy jeszcze jeden przykład, aby wyjaśnić tę koncepcję.

### Przykład: Wydrukuj liczby nieparzyste za pomocą Domknięcia Golanga

````python
def calculate():
    num = 1
    def inner_func():
        nonlocal num
        num += 2
        return num
    return inner_func

# wywołujemy funkcję zewnętrzną
odd = calculate()

# wywołujemy funkcję wewnętrzną
print(odd())
print(odd())
print(odd())

# ponownie wywołujemy funkcję zewnętrzną
odd2 = calculate()
print(odd2())
````

W powyższym przykładzie
```
odd = calculate()
```

Ten kod wykonuje funkcję zewnętrzną cal() i zwraca zamknięcie liczby nieparzystej. T

Właśnie dlatego możemy uzyskać dostęp do zmiennej num metody obliczania () nawet po wykonaniu funkcji zewnętrznej.

## Dekoratory Pythona

W Pythonie dekorator to wzorzec projektowy, który pozwala modyfikować funkcjonalność funkcji poprzez umieszczenie jej w innej funkcji.

Funkcja zewnętrzna nazywana jest dekoratorem, który przyjmuje pierwotną funkcję jako argument i zwraca jej zmodyfikowaną wersję.

### Mała powtórka

#### Funkcja zagnieżdżona
Możemy zawrzeć jedną funkcję wewnątrz drugiej, co nazywamy funkcją zagnieżdżoną. Na przykład,

````python
def outer(x):
    def inner(y):
        return x + y
    return inner

add_five = outer(5)
result = add_five(6)
print(result)
````

Tutaj utworzyliśmy funkcję inner() wewnątrz funkcji external().

#### Przekaż funkcję jako argument
Możemy przekazać funkcję jako argument do innej funkcji w Pythonie. Na przykład,

````python
def add(x, y):
    return x + y

def calculate(func, x, y):
    return func(x, y)

result = calculate(add, 4, 6)
print(result)
````

W powyższym przykładzie funkcja calculate() przyjmuje jako argument funkcję. Podczas wywoływania funkcji obliczania() jako argument przekazujemy funkcję add().

W funkcji calculate() argumenty: func, x, y stają się odpowiednio add, 4 i 6.

I stąd funkcja func(x, y) staje się add(4, 6), która zwraca 10.

#### Zwróć funkcję jako wartość
W Pythonie możemy również zwrócić funkcję jako wartość zwracaną. Na przykład,

````python
def greeting(name):
    def hello():
        return "Hello, " + name + "!"
    return hello

greet = greeting("Atlantis")
print(greet())  # wyświetli "Hello, Atlantis!"
````

W powyższym przykładzie instrukcja return hello zwraca wewnętrzną funkcję hello(). Ta funkcja jest teraz przypisana do zmiennej greet.

Dlatego też, gdy wywołujemy greet() jako funkcję, otrzymujemy dane wyjściowe.

### Dekoratory Pythona
Jak wspomniano wcześniej, dekorator Pythona to funkcja, która przyjmuje funkcję i zwraca ją, dodając pewną funkcjonalność.

W rzeczywistości każdy obiekt, który implementuje specjalną metodę __call__() jest określany jako wywoływalny. Tak więc, w najbardziej podstawowym sensie, dekorator to wywoływalny obiekt, który zwraca wywoływalny obiekt.

Zasadniczo dekorator przyjmuje funkcję, dodaje pewną funkcjonalność i zwraca ją.



````python
def make_pretty(func):
    def inner():
        print("I got decorated")
        func()
    return inner


def ordinary():
    print("I am ordinary")
````

Tutaj stworzyliśmy dwie funkcje:

* *ordinary()*, która wyświetla „Jestem zwyczajny”
* *make_pretty()*, która jako argument przyjmuje funkcję i ma zagnieżdżoną funkcję o nazwie *inner()* oraz zwraca funkcję wewnętrzną.

Wywołujemy funkcję *ordinary()* normalnie, więc otrzymujemy wynik „*Jestem zwyczajny*”. Teraz nazwijmy to za pomocą funkcji dekoratora.

````python
def make_pretty(func):
    # definiujemy funkcję wewnętrzną
    def inner():
        # dodajemy dodatkowe zachowanie do funkcji z dekoratorem
        print("I got decorated")

        # wywołujemy oryginalną funkcję
        func()
    # zwracamy wewnętrzną funkcję
    return inner

# definiujemy zwykłą funkcję
def ordinary():
    print("I am ordinary")

# dekorujemy zwykłą funkcję
decorated_func = make_pretty(ordinary)

# wywołujemy funkcję z dekoratorem
decorated_func()
````

W powyższym przykładzie make_pretty() jest dekoratorem. Zwróć uwagę na kod,

```
decorated_func = make_pretty(ordinary)
```

* Przekazujemy teraz funkcję common() jako argument funkcji *make_pretty()*.

* Funkcja make_pretty() zwraca funkcję wewnętrzną i jest teraz przypisana do zmiennej Decorated_func.
```
decorated_func()
```

Tutaj faktycznie wywołujemy funkcję inner(), w której robimy print

#### @ Symbol Z Dekoratorem
Zamiast przypisywać wywołanie funkcji do zmiennej, Python zapewnia znacznie bardziej elegancki sposób osiągnięcia tej funkcjonalności za pomocą symbolu @. Na przykład,

````python
def make_pretty(func):

    def inner():
        print("I got decorated")
        func()
    return inner

@make_pretty
def ordinary():
    print("I am ordinary")

ordinary()
````

W tym przypadku funkcja ordinary() jest ozdobiona dekoratorem make_pretty() przy użyciu składni @make_pretty, co jest równoważne wywołaniu zwyczajnego = make_pretty(ordinary).

### Dekorowanie funkcji za pomocą parametrów
Powyższy dekorator był prosty i działał tylko z funkcjami, które nie miały żadnych parametrów. Co by było, gdybyśmy mieli funkcje, które przyjmowałyby parametry takie jak:

````python
def divide(a, b):
    return a/b
````

Ta funkcja ma dwa parametry, a i b. Wiemy, że jeśli przekażemy b jako 0, wystąpi błąd.

Teraz stwórzmy dekorator, który sprawdzi ten przypadek, który spowoduje błąd.

````python
def smart_divide(func):
    def inner(a, b):
        print("I am going to divide", a, "and", b)
        if b == 0:
            print("Whoops! cannot divide")
            return

        return func(a, b)
    return inner

@smart_divide
def divide(a, b):
    print(a/b)

divide(2,5)

divide(2,0)
````

Tutaj, kiedy wywołujemy funkcję divide() z argumentami (2,5), zamiast niej wywoływana jest funkcja inner() zdefiniowana w dekoratorze smart_divide().

Ta funkcja inner() wywołuje oryginalną funkcję divide() z argumentami 2 i 5 i zwraca wynik, który wynosi 0,4.

Podobnie, gdy wywołujemy funkcję divide() z argumentami (2,0), funkcja inner() sprawdza, czy b jest równe 0 i wyświetla komunikat o błędzie przed zwróceniem Brak.


### Łączenie dekoratorów w Pythonie
W Pythonie można połączyć wiele dekoratorów.

Aby stworzyć łańcuch dekoratorów w Pythonie, możemy zastosować wiele dekoratorów do jednej funkcji, umieszczając je jeden po drugim, przy czym najbardziej wewnętrzny dekorator jest stosowany jako pierwszy.

````python
def star(func):
    def inner(*args, **kwargs):
        print("*" * 15)
        func(*args, **kwargs)
        print("*" * 15)
    return inner


def percent(func):
    def inner(*args, **kwargs):
        print("%" * 15)
        func(*args, **kwargs)
        print("%" * 15)
    return inner


@star
@percent
def printer(msg):
    print(msg)

printer("Hello")
````

@percent

    print(msg)

jest równoważna

    print(msg)
```

Kolejność, w jakiej łączymy dekoratory, ma znaczenie. Gdybyśmy odwrócili kolejność, np.
```

### Dekorator @property
W Pythonie właściwość() jest wbudowaną funkcją, która tworzy i zwraca obiekt właściwości. Składnia tej funkcji to:
```
property(fget=None, fset=None, fdel=None, doc=None)
```

Tutaj,

* fget to funkcja pobierająca wartość atrybutu
* fset to funkcja ustawiająca wartość atrybutu
* fdel to funkcja usuwająca atrybut
* doc to ciąg znaków (jak komentarz)

Jak widać z implementacji, te argumenty funkcji są opcjonalne.

Obiekt właściwości ma trzy metody, getter(), setter() i deleter() do określania fget, fset i fdel w późniejszym czasie. Oznacza to, że linię:

```
temperature = property(get_temperature,set_temperature)
```

można podzielić jako:

```
# make empty property
temperature = property()

# assign fget
temperature = temperature.getter(get_temperature)

# assign fset
temperature = temperature.setter(set_temperature)
```

Te dwa fragmenty kodu są równoważne.

Programiści zaznajomieni z dekoratorami Pythona mogą rozpoznać, że powyższą konstrukcję można zaimplementować jako dekoratory.

Nie możemy nawet zdefiniować nazw get_temperature i set_temperature, ponieważ są one niepotrzebne i zanieczyszczają przestrzeń nazw klas.

W tym celu ponownie używamy nazwy temperatury podczas definiowania naszych funkcji pobierających i ustawiających. Przyjrzyjmy się, jak zaimplementować to jako dekorator:

````python
# Korzystanie z @property
class Celsius:
    def __init__(self, temperature=0):
        self.temperature = temperature

    def to_fahrenheit(self):
        return (self.temperature * 1.8) + 32

    @property
    def temperature(self):
        print("Getting value...")
        return self._temperature

    @temperature.setter
    def temperature(self, value):
        print("Setting value...")
        if value < -273.15:
            raise ValueError("Temperature below -273 is not possible")
        self._temperature = value


# tworzenie obiektu
human = Celsius(37)

print(human.temperature)

print(human.to_fahrenheit())

coldest_thing = Celsius(-300)
````

Powyższa implementacja jest prosta i wydajna. Jest to zalecany sposób korzystania z nieruchomości.

## RegEx

### RegEx w Pythonie
Python ma moduł o nazwie re do pracy z wyrażeniami regularnymi. Aby z niego skorzystać, musimy zaimportować moduł.
```
import re
```
Moduł definiuje kilka funkcji i stałych do pracy z RegEx.

#### re.findall()
Metoda re.findall() zwraca listę stringów zawierających wszystkie dopasowania.

````python
# Program to extract numbers from a string

import re

string = 'hello 12 hi 89. Howdy 34'
pattern = '\d+'

result = re.findall(pattern, string)
print(result)
````

Jeśli wzorzec nie zostanie znaleziony, re.findall() zwróci pustą listę.

#### re.split()
Metoda re.split dzieli string, w którym występuje dopasowanie, i zwraca listę ciągów, w których wystąpiły podziały.



````python
import re

string = 'Twelve:12 Eighty nine:89.'
pattern = '\d+'

result = re.split(pattern, string)
print(result)
````

Jeśli wzorzec nie zostanie znaleziony, re.split() zwraca listę zawierającą oryginalny łańcuch.

Możesz przekazać argument maxsplit do metody re.split() . Jest to maksymalna liczba podziałów, które wystąpią.

````python

import re

string = 'Twelve:12 Eighty nine:89 Nine:9.'
pattern = '\d+'

maxsplit = 1
# split only at the first occurrence
result = re.split(pattern, string, maxsplit)
print(result)
````

Nawiasem mówiąc, domyślną wartością maxsplit jest 0; czyli wszystkie możliwe podziały.

#### re.sub()
Składnia re.sub() jest następująca:
```
re.sub(pattern, replace, string)
```
Metoda zwraca ciąg znaków, w którym dopasowane wystąpienia są zastępowane treścią zmiennej zamiany.

````python
# Program do usuwania wszystkich białych znaków
import re

# string na kilka linijek
string = 'abc 12\
de 23 \n f45 6'

# dopasowanie do wszytkich białych znaków
pattern = '\s+'

# pusty string
replace = ''

new_string = re.sub(pattern, replace, string)
print(new_string)
````

Jeśli wzorzec nie zostanie znaleziony, re.sub() zwraca oryginalny łańcuch.

Możesz przekazać count jako czwarty parametr do metody re.sub() . Jeśli zostanie pominięty, wynikiem będzie 0. Spowoduje to zastąpienie wszystkich wystąpień.

````python
import re

# multiline string
string = 'abc 12\
de 23 \n f45 6'

# matches all whitespace characters
pattern = '\s+'
replace = ''

new_string = re.sub(r'\s+', replace, string, 1)
print(new_string)
````

#### re.subn()
Funkcja re.subn() jest podobna do funkcji re.sub() z tą różnicą, że zwraca krotkę złożoną z 2 elementów zawierających nowy łańcuch i liczbę dokonanych podstawień.

````python
# Program do usuwania wszystkich białych znaków
import re

# string na kilka linijek
string = 'abc 12\
de 23 \n f45 6'

# dopasowanie do wszytkich białych znaków
pattern = '\s+'

# pusty string
replace = ''

new_string = re.subn(pattern, replace, string)
print(new_string)

# Output: ('abc12de23f456', 4)
````

####re.search()
Metoda re.search() przyjmuje dwa argumenty: wzorzec i ciąg znaków. Metoda szuka pierwszej lokalizacji, w której wzorzec RegEx tworzy dopasowanie do łańcucha.

Jeśli wyszukiwanie zakończy się pomyślnie, re.search() zwróci obiekt dopasowania; jeśli nie, zwraca None.
```
match = re.search(pattern, str)
```

````python

import re

string = "Python is fun"

# sprawdź, czy na początku jest „Python”.
match = re.search('\APython', string)

if match:
  print("pattern found inside the string")
else:
  print("pattern not found")

# Output: pattern found inside the string
````

Tutaj dopasowanie zawiera obiekt dopasowania

#### Dopasowaie obiektu
Możesz uzyskać metody i atrybuty obiektu dopasowania za pomocą funkcji dir().

Niektóre z powszechnie używanych metod i atrybutów obiektów dopasowania to:

##### match.group()
Metoda group() zwraca część ciągu znaków, w której występuje dopasowanie.

````python
import re

string = '39801 356, 2102 1111'

# Liczba trzycyfrowa, po której następuje spacja i liczba dwucyfrowa
pattern = '(\d{3}) (\d{2})'

# zmienna match zawiera obiekt Match.
match = re.search(pattern, string)

if match:
  print(match.group())
else:
  print("pattern not found")
````

Tutaj zmienna dopasowania zawiera obiekt dopasowania.

Nasz wzorzec (\d{3}) (\d{2}) ma dwie podgrupy (\d{3}) i (\d{2}). Możesz uzyskać część ciągu tych podgrup w nawiasach. Oto jak:
```
>>> match.group(1)
'801'

>>> match.group(2)
'35'
>>> match.group(1, 2)
('801', '35')

>>> match.groups()
('801', '35')
```

##### match.start(), match.end() oraz match.span()
Funkcja start() zwraca indeks początku dopasowanego podłańcucha. Podobnie metoda end() zwraca indeks końcowy dopasowanego podłańcucha.

```
>>> match.start()
2
>>> match.end()
8
```
Funkcja span() zwraca krotkę zawierającą indeks początkowy i końcowy dopasowanej części.
```
>>> match.span()
(2, 8)
```

##### match.re oraz match.string
Atrybut re dopasowanego obiektu zwraca obiekt wyrażenia regularnego. Podobnie atrybut string zwraca przekazany ciąg.
```
>>> match.re
re.compile('(\\d{3}) (\\d{2})')

>>> match.string
'39801 356, 2102 1111'
```

#### Używanie przedrostka r przed RegEx
Gdy prefiks r lub R jest używany przed wyrażeniem regularnym, oznacza to nieprzetworzony ciąg znaków. Na przykład '\n' to nowa linia, podczas gdy r'\n' oznacza dwa znaki: ukośnik odwrotny \, po którym następuje n.

Backlash \ służy do ucieczki przed różnymi znakami, w tym wszystkimi metaznakami. Jednak użycie przedrostka r powoduje, że \ traktuje się jak normalny znak.



````python
import re

string = '\n and \r are escape sequences.'

result = re.findall(r'[\n\r]', string)
print(result)
````
