````python
# np. chcemy obliczyć sumę liczb w zakresie
# tradycyjnie:
def sum(n):
    total = 0
    for index in range(n+1):
        total += index

    return total


result = sum(10)
print(result)

# rekurencyjnie
def sum_rec(n):
  if n > 0:
    return n + sum_rec(n-1)
  return 0
result_rec = sum_rec(10)
print(result_rec)
````

````python
# Teraz możemy przerobić naszą funkcję silnii na funkcję rekurencyjną
def factorial(x):
    """Piszemy funkcję rekurencyjną
    do obliczenia x!"""

    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))


num = 3
print(f"{num}! jest równe {factorial(num)}")
````

### Global, local i nonlocal

W Pythonie zmienne dzielimy na lokalne, globalne i nielokalne w zależności od tego w jakim "obszarze" kodu są dostępne.

#### Local
Zmienne lokalne istnieją tylko wewnątrz funkcji, nie możemy z nich skorzystać poza.

````python
def greet():

    # zmienna lokalna
    message = 'Hello'

    print('Local', message)

greet()

# spróbujmy skorzystać z tej zmiennej poza funkcją
print(message)
````

#### Global
Zmienne zadeklarowane poza funkcją nazywami globalnymi, ponieważ są dostępne zarówno wewnątrz jak i poza funkcją.

````python
# deklarujemy zmienną globalną
message = 'Hello'

def greet():
    print('Local', message)

greet()
print('Global', message)
````

#### Nonlocal

Zmienne nielokalne nie istnieją ani lokalnie ani globalnie, korzystamy z nich w funkcjach wewnętrznych.

````python
# funkcja zewnętrzna
def outer():
    message = 'local'

    # funkcja zagnieżdżona
    def inner():

        # deklarujemy zmienną nonlocal
        nonlocal message

        message = 'nonlocal'
        print("inner:", message)

    # inner() # wywołana wcześniej nadpisze nam zmienną message
    print("outer:", message)
    inner()

outer()
````

### Słowo klucz **global**

Słowo klucz **global** pozwala na modyfikacje zmiennej globalnej wewnątrz funkcji.

````python
c = 1 # zmienna globalna

def add():
    print(c)

add()
````

````python
c = 1 # zmienna globalna

def add():

     # zwiększamy c o 2, ale dostaniemy błąd, bo nie możemy zmienić zmiennej globalnej wewnątrz funkcji
    c = c + 2

    print(c)

add()
````

````python
c = 1

def add():

    # dopiero jak skorzystamy ze słowa kluczowego global, poniższe staje się możliwe
    global c

    # zwiększamy c o 2
    c = c + 2

    print(c)

add()
````

````python
def outer_function():
    num = 20

    def inner_function():
        global num # tworzymy zmienną globalną
        num = 25 # te zmiany będą widocznie globalnie, ale nie lokalnie

    print("Before calling inner_function(): ", num) # zwróci 20 bo wciąć jesteśmy w zakresie lokalnym
    inner_function() # to zmieni zmienną globalną
    print("After calling inner_function(): ", num) # zwróci 20 bo wciąć jesteśmy w zakresie lokalnym

outer_function()
print("Outside both function: ", num)
````

### Dodawanie do listy
Możemy dodwać elementy do listy na kilka sposobów:

1. Metodą append()

Metoda append() dodaje element do listy na końcu (dokleja).

````python
numbers = [21, 34, 54, 12]

print("Before Append:", numbers)

# zastosowanie append
numbers.append(32)

print("After Append:", numbers)
````

Metodą append dodaliśmy liczbę 32 do końca listy. Wszystko w jednej linijce bez nadpisywania obiektu:
```
numbers.append(32)
```




2. Metoda extend()

Metodą extend()dodajemy elementy jednej listy do drugiej.

````python
prime_numbers = [2, 3, 5]
print("List1:", prime_numbers)

even_numbers = [4, 6, 8]
print("List2:", even_numbers)

# zŁączenie dwóch list
prime_numbers.extend(even_numbers)

print("List after append:", prime_numbers) 
````

W powyższym kodzie w linijce `prime_numbers.extend(even_numbers)` rozszerzamy listę prime_numbers o zawartość listy even_numbers.

### Zmiana elementów listy
Listy w Pythonie są mutowalne - możemy je zmieniać (edytować).
Możemy zmieniać obiekty w liście poprzez nadpisanie wartości operatorem =.

### Usuwanie z listy
1. Metodą del()

Możemy używać del do usuwania elementu po indeksie.

````python
languages = ['Python', 'Swift', 'C++', 'C', 'Java', 'Rust', 'R']

# usunięcie 2 elementu
del languages[1]
print(languages) # ['Python', 'C++', 'C', 'Java', 'Rust', 'R']

# usunięcie ostatniego elementu
del languages[-1]
print(languages) # ['Python', 'C++', 'C', 'Java', 'Rust']

# usunięcie 2 elementów
del languages[0 : 2]  # ['C', 'Java', 'Rust']
print(languages)
````

2. Metodą remove()

Ta metoda usuwa ostatni element z listy, chyba, że przekażemy inny argument.

````python
languages = ['Python', 'Swift', 'C++', 'C', 'Java', 'Rust', 'R']

# usuwamy 'Python'
languages.remove('Python')

print(languages) # ['Swift', 'C++', 'C', 'Java', 'Rust', 'R']
````

## Krotki (Tuples)

Krotki są podobne do list, ale są niemutowalne. To oznacza, że po utworzeniu nie można już zmienić ich zawartości.

### Tworzenie krotki
Krotkę tworzymy poprzez nawiasy () i rozdzielenie obiektów przecinkami. Nawiasy są opcjonalne, ale warto je stosować.

Krotka może mieć dowolną liczbę elementow różnych typów (integer, float, list, string, etc.).

````python
# Pusta krotka
my_tuple = ()
print(my_tuple)

# Krotka liczb
my_tuple = (1, 2, 3)
print(my_tuple)

# Krotka z danymi mieszanymi
my_tuple = (1, "Hello", 3.4)
print(my_tuple)

# Krotka zagnieżdżona
my_tuple = ("mouse", [8, 4, 6], (1, 2, 3))
print(my_tuple)
````

Krotki moż na również tworzyć bez nawiasów.

```
my_tuple = 1, 2, 3
my_tuple = 1, "Hello", 3.4
```



Utworzenie krotki jednoelementowej nie jest takie oczywiste. Jeśli wpiszemy wartość w samych nawiasach nie dostaniemy krotki ale np. string.

Potrzebujemy przecinka aby wskazać, że to ma być krotka

```
var1 = ("Hello") # string
var2 = ("Hello",) # tuple
```
Skorzystamy z funkcji() aby sprawdzić, do której klasy należy utworzony obiekt.


````python
var1 = ("hello")
print(type(var1))  # <class 'str'>

# tworzenie krotki jednoelementowej
var2 = ("hello",)
print(type(var2))  # <class 'tuple'>

# opcjonalne nawiasy
var3 = "hello",
print(type(var3))  # <class 'tuple'>
````

Powużej:

* ("hello") jest ciągiem znaków, zatem type() zwraca klasę str zmiennej var1 i.e. <class 'str'>
* ("hello",) oraz "hello", są krotkami, zatem type() zwraca klasę tupple zmiennej var2 i var3 i.e. <class 'tuple'>

### Dlaczego wyieramy krotki zamiast list w Pythonie? (Nie zawsze 😀)


Krotki są podobne do list i są wykorztstywane w wielu podobnych sytuacjach.

Ale krotki mają przewagę w kilku aspektach:

* Korzystamy z krotek do danych o wielu typach, a list do danych o jednym typie.
* Ponieważ krotki są niemutowalne, ich iteracja jest szybsza niż list i ich zastosowanie sprzyja optymalizacji kodu. 
* Krotki, które są niemutowalne, mogą służyć jako klucze w słowniku, w przeciwieństwie do list.
* Jeżeli twoje dane nie ulegną zmianie, krotki zapewniają lepszą ochronę przed modyfikacjami. 

## Zbiory (Sets)

Zbiór to kolekcja unikalnych danych. Nie zawierają duplikatów.

Mogą być wykorzystane do przechowywania np. esek studenckich, ponieważ te nie mogą się powtarzać.

image.png

### Tworzenie zbioru

Tworzymy za pomocą nawiasów {} io oddzielamy dane przecinkiem. 
Zbiór może składać się z dowolnej liczby obiektów i typów mieszanych, ale nie może zawierać obiektów mutowalnych jak listy, zbiory czy słowniki.

````python
# zbiór liczb
student_id = {112, 114, 116, 118, 115}
print('Student ID:', student_id)

# zbiór stringów
vowel_letters = {'a', 'e', 'i', 'o', 'u'}
print('Vowel Letters:', vowel_letters)

# zbiór mieszany
mixed_set = {'Hello', 101, -2, 'Bye'}
print('Set of mixed data types:', mixed_set)
````

Ważne: Zbiory to kolekcje nieuporządkowane, jeśli wyświetlimy zbiór możemy dostać inną kolejność elementów, niż zadeklarowaliśmy. 

### Tworzenie pustego zbiory
Utworzenie pustego zbioru nie jest takie oczywiste. Puste nawiasy {} utworzą nam obiekt pustego słownika. 
Musimy użyć funkcji set() bez podania argumentów.


````python
# tworzymy pusty zbiór
empty_set = set()

# tworzymy pusty słownik
empty_dictionary = { }

# sprawdzamy typ
print('Data type of empty_set:', type(empty_set))

print('Data type of empty_dictionary', type(empty_dictionary))
````

### Duplikaty w zbiorze
Spróbujmy dodać duplikaty do zbioru.

````python
numbers = {2, 4, 6, 6, 2, 8}
print(numbers)   # {8, 2, 4, 6}
````

Jak widzimy duplikaty zostaną zredukowane do unikalnych wartości.

### Dodawanie i edycja obiektów w zbiorze

Zbiory są mutowalne, ale ponieważ są nieuporządkowane, nie można stosować indeksacji ani wycinania.

### Dodawanie elementu
Metoda add() pozwala na dodanie elementu do zbioru.

````python
numbers = {21, 34, 54, 12}

print('Initial Set:',numbers)

numbers.add(32)

print('Updated Set:', numbers) 
````

Linijka

```
numbers.add(32)
```
dodaje 32 do naszego zbioru.


### Edycja zbioru
Metoda update() pozwala powiększyć zbiór o elementy z innej kolekcji, np. listy.

````python
companies = {'Lacoste', 'Ralph Lauren'}
tech_companies = ['apple', 'google', 'apple']

companies.update(tech_companies)

print(companies)

# Output: {'google', 'apple', 'Lacoste', 'Ralph Lauren'}
````

Wszystkie unikalne wartości z tech_companies zostały dodane do companies.

### Usuwanie elemtnuy
Metoda discard() usuwa wskazany element w zbiorze.

````python
languages = {'Swift', 'Java', 'Python'}

print('Initial Set:',languages)

# usuniemy 'Java' ze zbioru
removedValue = languages.discard('Java')

print('Set after remove():', languages)
````

###Operacje na zbiorach

Python zawiera wbudowane funkcje do matematycznych operacji na zbiorach:

1. Suma zbiorów
Suma zbiorów A i B to wszystkie elementy w zbiorach A i B.



image.png

Możemy wykorzystać operator | albo metodę union() aby wykonać opercję sumy zbiorów.

````python
# pierwszy zbiór
A = {1, 3, 5}

# drugi zbiór
B = {0, 2, 4}

# suma za pomocą |
print('Union using |:', A | B)

# suma za pomocą union()
print('Union using union():', A.union(B)) 
````

Ważne: Operacje A|B oraz union() są równoważne A ⋃ B.

2. Część wspólna

Część wspólna zbiorów A i B to elementy występujące w obu zbiorach A i B.

image.png

Możemy wykorzystać operator & lub metodę intersection() aby wykonać operację części wspólnej.

````python
A = {1, 3, 5}

B = {1, 2, 3}

# część wspólna z &
print('Intersection using &:', A & B)

# część wspólna z intersection()
print('Intersection using intersection():', A.intersection(B)) 
````

Wznea: Operacje A&B oraz intersection() są równoważne A ⋂ B.

3. Różnica zbiorów

Różnica zbiorów A i B zawiera elementy ze zbioru A, których nie ma w zbiorze B.

image.png

Używamy operatora - operator albo metody difference() aby obliczyć różnicę zbiorów.

````python
A = {2, 3, 5}

B = {1, 2, 6}

# różnica z &
print('Difference using &:', A - B)

# różnica z difference()
print('Difference using difference():', A.difference(B)) 
````

Ważne: A - B oraz A.difference(B) są równoważne A - B.

4. Różnica symetryczna zbiorów
 
Różnica symetryczna zbiorów to wszystkie elementy ze zbiorów A i B poza częścią wspólną. 

image.png

Możemy wykorzystać operator ^ albo metodę symmetric_difference() aby wykonać operachę różnicy symetrycznej zbiorów.

````python
A = {2, 3, 5}

B = {1, 2, 6}

# operacja z ^
print('using ^:', A ^ B)

# operacja z symmetric_difference()
print('using symmetric_difference():', A.symmetric_difference(B)) 
````

````python
# napisz funkcję, która będzie WYŚWIETLAŁA ciąg Fibonnaciego o długości podanej w argumencie
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        print(a, end=' ')
        a,b = b, a + b
    print()


fib(10)
````

````python
# napisz funkcję, która ZWRACA silnię podanej liczby. Upewnij się, że silnia jest możliwa, w przeciwnym wypadku wyświetl komunikat o błędzie i nie zwracaj nic
def silnia(liczba):
    if liczba < 0:
        print("Błąd - liczba nie moze byc ujemna do silni")
        return None
    result = 1
    for i in range(1,liczba + 1):
        result *= i
    return result

silnia(4)
````

````python
# napisz funkcję, która będzie przyjmowała dowolną liczbę liczb i zwracała ich iloczyn
def mnozenie(*liczby):
    result = 1;
    for liczba in liczby:
        result *= liczba
    return result

mnozenie(2,3,4)
````

````python
# napisz funkcję, która będzie przyjmowała dowolną liczbę liczb i zwracała ich iloczyn i iloraz w jednej funkcji
def mnozenie_dzielenie(*liczby):
    if not liczby:
        return None,None
    iloczyn,iloraz = 1,liczby[0]
    for liczba in liczby:
        iloczyn *= liczba;
    for liczba in liczby[1:]:
        if liczba == 0:
            print("Dzielenie przez zero 0")
            return iloczyn, None
        iloraz /= liczba;
    return iloczyn, iloraz

mnozenie_dzielenie(5,1,2)
````

````python
# napisz funkcję, która będzie przyjmowała dwie liczby oraz string, z informacją jaka to ma być operacja matematyczna - do wyboru: dodawanie, odejmowanie, mnożenie i dzielenie
# jeśli nie podano operacji, domyślnie będzie to dodawanie
# dodatkowo postaraj się skorzystać z konstrukcji match case
# ma zwracać wynik wybranej operacji

def operacje(x,y,op = "dodawanie"):
    match op:
        case "dodawanie": return x + y
        case "odejmowanie": return x - y
        case "mnożenie": return x * y
        case "dzielenie":
            if y == 0:
                print("Dzielenie przez zero")
                return None
            return x / y
        case _:
            print("Nie ma takiej operacji")
            return None
        
operacje(12,3,"dzielenie")
# operacje(2,4)
````

````python
# odwróć kolejność elementów w krotce
tuple1 = (10, 20, 30, 40, 50)

rev_tuple1 = tuple1[::-1]
print(rev_tuple1)
````

````python
# wyciagnij z krotki wartość 20
tuple1 = ("Orange", [10, 20, 30], (5, 15, 25))
val = tuple1[1][1]
print(val)
````

````python
# sprawdź czy wszystkie elementy w krotce są identyczne

tuple1 = (45, 45, 45, 45)
is_same = len(set(tuple1)) == 1
print(is_same)
````
